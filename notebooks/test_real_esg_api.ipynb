```python
import os
import requests
import json
import pandas as pd

API_KEY = os.environ.get("ESG_API_KEY", "")
BASE = "https://esg-data-api.p.rapidapi.com/v1/screen"
headers = {
    "x-rapidapi-key": API_KEY,
    "x-rapidapi-host": "esg-data-api.p.rapidapi.com"
}

all_results = []
offset = 0
while True:
    r = requests.get(BASE, headers=headers, params={
        "min_esg": 0,
        "limit": 100,
        "offset": offset
    })
    parsed = r.json()
    results = parsed.get('results', [])
    all_results.extend(results)
    print(f"Offset {offset}: {len(results)} geholt, gesamt: {len(all_results)}")
    if len(results) < 100:
        break
    offset += 100

print(f"\nTotal: {len(all_results)} Unternehmen")
df = pd.json_normalize(all_results)
df[['ticker', 'name', 'country', 'sector', 'scores.overall']].head(20)
```

```python
import os
# === Alle Endpoints testen ===
import requests, json

API_KEY = os.environ.get("ESG_API_KEY", "")
BASE = "https://esg-data-api.p.rapidapi.com"
headers = {
    "x-rapidapi-key": API_KEY,
    "x-rapidapi-host": "esg-data-api.p.rapidapi.com"
}

def test_endpoint(path, label):
    r = requests.get(f"{BASE}{path}", headers=headers)
    print(f"\n{'='*60}")
    print(f"{label}: {path}")
    print(f"Status: {r.status_code}")
    if r.status_code == 200:
        print(json.dumps(r.json(), indent=2)[:1500])
    else:
        print(r.text[:500])
```

```python
# 1. Company Details — Stammdaten eines Unternehmens
test_endpoint("/v1/companies/AAPL", "Company Details Apple")
```
